```python
places = []
radius = 20*1000
for lat, lon, country, capital in pc_df[['CapitalLatitude','CapitalLongitude','CountryName','CapitalName']].values:
    r = re.get(f'https://api.opentripmap.com/0.1/ru/places/radius?radius={radius}&lon={lon}&lat={lat}&rate=3h&apikey=5ae2e3f221c38a28845f05b63e35ac766a56ae6600d39b36da2dc4a4')
    data = r.json()
    for item in data['features']:
        place = [country, capital,
                 item['properties']['name'],
                 item['properties']['rate'],
                 item['properties']['xid'],
                 item['properties']['kinds'],
                 item['geometry']['coordinates'][0],
                 item['geometry']['coordinates'][1]]
        places.append(place)
```

```python
pl = pd.DataFrame(places, columns=['country', 'capital', 'name','rate', 'xid', 'kinds', 'lat', 'lon'])
```

```python
pl = pd.read_csv('places.csv')
```

```python
pl['rate_norm'] = [0]*pl.shape[0]
pl['description'] = [0]*pl.shape[0]
pl['image'] = [0]*pl.shape[0]
```

```python
pl[['rate_norm','description','image']] = [[0]*pl.shape[0],[0]*pl.shape[0],[0]*pl.shape[0]]
```

```python
pl.xid[:20]
```

```python
places_info = []
for xid in tqdm(pl.xid):
    r = re.get(f'http://api.opentripmap.com/0.1/ru/places/xid/{xid}?apikey=5ae2e3f221c38a28845f05b63e35ac766a56ae6600d39b36da2dc4a4')
    item = r.json()
    info = [xid,
             item['rate']if 'rate' in item else None,
             item['wikipedia_extracts']['text'] if 'wikipedia_extracts' in item else None,
             item['image'] if 'image' in item else None
            ]
    places_info.append(info)
        
```

```python
info_df = pd.DataFrame(places_info, columns=['xid','rate_norm', 'wikipedia_extracts', 'image'])
```

```python
info_df.to_csv('info_df.csv', index = False)
```

```python
full_df = pd.merge(pl, info_df,
how='inner', on='xid')
```
